```python
data1A=[10,20,3,5,1,3,2]
data1B=[15,31,6,7,2,8,1]
inputValues=[x for x in range (-10,10)] 
```

```python
def sigmoidFunction(x): 
    return 1/(1+math.exp(-x)) 
```

```python
outputValues=[sigmoidFunction(x) for x in inputValues]
```

```python
plt.plot(data1A,data1B,label="my graph")
plt.plot(inputValues,outputValues,color='blue',marker='o',label="sigmoid") 
plt.axis([-10,10,0,1]) 
```
